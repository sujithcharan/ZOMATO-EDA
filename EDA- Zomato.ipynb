```python
rating_data =  pd.DataFrame(final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns = {0:"rating count"}))
```

```python
rating_data
# creating new table dedicated to rating data 
```

```python
plt.rcParams["figure.figsize"] = (12,6)
# y_axis = rating_data["rating count"][1:]
fig, ax = plt.subplots()
fig.patch.set_facecolor('#2d2d2d')  # Dark grey background for the figure
ax.set_facecolor('#333335') 
sns.barplot(x = "Aggregate rating", y = "rating count",hue ="Rating color",data = rating_data,color="g",palette = ["white","red","orange","yellow","green","green"])
# graph of frequency of data values 
```

```python
sns.countplot(x = "Rating color",data = rating_data,palette = ["white","red","orange","yellow","green","green"])
# count plot for frequency of rating color 
```
